```python
import numpy as np
from glob import glob
import os
import random
```

```python
def train_test_split_from_directory(root_path, train_size=0.8):
    train, test = [], []

    images = glob(os.path.join(root_path, '*.jpg')) + \
             glob(os.path.join(root_path, '*.png')) + \
             glob(os.path.join(root_path, '*.jpeg'))

    random.shuffle(images)
    split_idx = np.int32(train_size * len(images))
    train.extend(images[:split_idx])
    test.extend(images[split_idx:])

    return train, test
```
